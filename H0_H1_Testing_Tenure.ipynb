You might want to test if the average client tenure (how long they’ve been with Vanguard) of those engaging with the new process is the same as those engaging with the old process

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as st

%matplotlib inline
```

```python
# Data Cleaning
file_path_clients = "C:/Users/Srnzzz/Documents/GitHub/vanguard-ab-test/client_old_new_df.csv"
df_clients = pd.read_csv(file_path_clients)
df_clients['date_time'] = pd.to_datetime(df_clients['date_time'])
df_clients_clean = df_clients.dropna()
```

```python
df_clients_clean
```

```python
file_path_sample = "C:/Users/Srnzzz/Documents/GitHub/vanguard-ab-test/new_final_demo.csv"
df_sample = pd.read_csv(file_path_sample)
df_sample
```

```python
df_merged_age = df_clients_clean.merge(df_sample[['client_id', 'clnt_tenure_yr']], on='client_id', how='left')
df_merged_age
```

```python
df_no_test = df_merged_age[df_merged_age['Variation'] != 'Test']
df_no_test
```

```python
df_test = df_merged_age[df_merged_age['Variation'] == 'Test']
df_test
```

```python
df_no_test_tenure = df_no_test.groupby('client_id')['clnt_tenure_yr'].mean().reset_index()
df_no_test_tenure_list = df_no_test_tenure["clnt_tenure_yr"]
df_no_test_tenure_list
```

```python
df_test_tenure = df_test.groupby('client_id')['clnt_tenure_yr'].mean().reset_index()
df_test_tenure_list = df_test_tenure["clnt_tenure_yr"]
df_test_tenure_list
```

```python
# ttest for average duration
#H0: test_average_tenure == control_average_age
#H1: test_average_age != control_average_age

df_test_clean = df_test_tenure_list.dropna()
df_no_test_clean = df_no_test_tenure_list.dropna()

st.ttest_ind(df_test_clean,df_no_test_clean, alternative="two-sided") 
```
